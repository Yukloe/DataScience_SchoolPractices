## Distplot

displot est une fonction de la bibliothèque seaborn en Python qui permet de représenter graphiquement la distribution d'une variable quantitative. Il peut s'agir d'une distribution univariée (une seule variable) ou bivariée (deux variables).

Lorsqu'il est utilisé avec une seule variable, displot génère un histogramme et/ou un kernel density estimate (KDE) pour visualiser la distribution de la variable. Si deux variables sont fournies, displot peut également générer une representation de la distribution conditionnelle des deux variables sous forme de lignes de densité ou de heatmap.

L'utilisation de displot est très simple : il suffit de lui fournir la variable quantitive et il génère automatiquement le graphique correspondant. Il offre également de nombreuses options pour personnaliser la représentation de la distribution (par exemple, le nombre de bins, la forme du KDE, etc.).

Le diagramme de distribution montre la distribution d'un ensemble d'observations univariées.

```python
sns.distplot(tips['total_bill'])
```

```python
# Pour supprimer la couche kde et avoir uniquement l'histogramme.
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

## Jointplot

Jointplot est une fonction de la bibliothèque seaborn en Python qui permet de représenter graphiquement les relations entre deux variables quantitatives en utilisant deux techniques de visualisation différentes : un histogramme pour chacune des variables et une representation de la distribution conjointe des deux variables à l'aide d'un nuage de points ou d'une heatmap.

Le but de Jointplot est de permettre une exploration visuelle des liens et de la distribution des deux variables pour déterminer si elles sont liées et de quelle manière.

L'utilisation de Jointplot est très simple : il suffit de lui fournir les deux variables quantitatives et il génère automatiquement le graphique correspondant.

`jointplot()` Le jointplot() vous permet de faire correspondre deux diagrammes de distribution pour les données bivariées. Avec votre choix du paramètre à comparer :
* “scatter” 
* “reg” 
* “resid” 
* “kde” 
* “hex”

### Scatter 

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

### Hex 

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

### Regression 

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

## Pairplot

pairplot est une fonction de la bibliothèque seaborn en Python qui permet de représenter graphiquement les relations entre plusieurs variables quantitatives dans un seul graphique. Cela permet une exploration rapide et visuelle des liens entre les variables et de leur distribution individuelle.

pairplot génère un graphique qui comporte une grille de sous-graphiques, chacun représentant les relations entre deux variables différentes. Pour chaque paire de variables, un histogramme et/ou un nuage de points sont utilisés pour représenter leur distribution conjointe. Si plus de deux variables sont fournies, un triangle supérieur rempli de graphiques représentant les distributions des variables restantes est également généré.

L'utilisation de pairplot est très simple : il suffit de lui fournir les variables quantitatives et il génère automatiquement le graphique correspondant. Il offre également de nombreuses options pour personnaliser les graphiques générés (par exemple, la forme des points, la couleur, etc.).

Le pairplot tracera les relations paires à travers tout un dataframe (pour les colonnes numériques) et prend en charge un argument de couleur hue (pour les colonnes catégoriques).

Les principaux paramètres de pairplot sont :

data : le DataFrame Pandas qui contient les variables quantitatives que vous souhaitez représenter.

vars : une liste des noms de colonnes ou des variables quantitatives à inclure dans le graphique. Si ce paramètre n'est pas spécifié, toutes les colonnes numériques seront utilisées.

hue : le nom de la colonne ou de la variable à utiliser pour colorier les points dans les nuages de points.

palette : la palette de couleurs à utiliser pour les couleurs.

diag_kind : le type de graphique à utiliser pour les graphiques sur la diagonale (histogramme, KDE, etc.).

kind : le type de graphique à utiliser pour les graphiques hors de la diagonale (scatterplot, heatmap, etc.).

markers : les marqueurs à utiliser pour les points dans les graphiques de nuages de points.

plot_kws : un dictionnaire de paramètres supplémentaires à passer aux fonctions de traçage sous-jacentes (par exemple, la taille des points).

diag_kws : un dictionnaire de paramètres supplémentaires à passer aux graphiques de la diagonale.

hist_kws : un dictionnaire de paramètres supplémentaires à passer aux histogrammes.

kde : un booléen indiquant s'il faut afficher un estimateur de densité de noyau (KDE) pour les variables quantitatives.

Il est important de noter que pairplot offre de nombreuses autres options pour personnaliser le graphique généré. Vous pouvez trouver la documentation complète pour la fonction pairplot sur le site web de Seaborn.

```python
sns.pairplot(tips)
```

```python
# separate by sex and color 
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

## Rugplot

rugplots est une fonction de la bibliothèque seaborn en Python qui permet de représenter graphiquement la distribution d'une variable quantitative sous forme de lignes verticales. Les lignes verticales (ou "rugs") sont tracées à côté des observations pour indiquer la position relative des valeurs dans la distribution.

Le rugplot est similaire à un histogramme, mais il montre la distribution des données en détail en plaçant une ligne verticale pour chaque observation. Cela peut être utile pour visualiser la distribution de petits jeux de données ou pour ajouter une dimension supplémentaire à un autre graphique, comme un boxplot ou un histogramme.

L'utilisation de rugplot est très simple : il suffit de lui fournir la variable quantitive et il génère automatiquement le graphique correspondant. Il offre également de nombreuses options pour personnaliser la représentation des lignes verticales (par exemple, la couleur, la largeur, etc.).

Les principaux paramètres de rugplot sont :

a : la variable quantitative à représenter sous forme de lignes verticales.

height : la hauteur des lignes verticales.

axis : l'axe auquel les lignes verticales seront attachées (x ou y).

axis_index : l'index de l'axe à utiliser pour placer les lignes verticales (0 pour l'axe des x et 1 pour l'axe des y).

color : la couleur des lignes verticales.

linewidth : la largeur des lignes verticales.

alpha : la transparence des lignes verticales.

marker : le marqueur à utiliser pour les extrémités des lignes verticales.

markersize : la taille des marqueurs.

Il est important de noter que rugplot offre de nombreuses autres options pour personnaliser la représentation des lignes verticales. Vous pouvez trouver la documentation complète pour la fonction rugplot sur le site web de Seaborn.



```python
sns.rugplot(tips['total_bill'])
```

## kdeplot

kdeplot est une fonction de la bibliothèque seaborn en Python qui permet de représenter graphiquement la distribution d'une variable quantitative sous forme de courbe de densité. La courbe de densité est un estimateur de la fonction de densité de probabilité de la variable quantitative. Cela permet de visualiser de manière plus claire la forme de la distribution et les tendances sous-jacentes des données.

L'utilisation de kdeplot est très simple : il suffit de lui fournir la variable quantitive et il génère automatiquement le graphique correspondant. Il offre également de nombreuses options pour personnaliser la représentation de la courbe de densité (par exemple, la couleur, la forme, etc.).

Les courbes de densité sont souvent utilisées en conjonction avec d'autres graphiques, comme des histogrammes ou des boxplots, pour visualiser la distribution des données de manière plus complète. En utilisant une courbe de densité, vous pouvez facilement identifier les modes, les skewness et les outliers dans vos données, ce qui peut être très utile pour la compréhension et l'analyse des données.

Les principaux paramètres de kdeplot sont :

data : la variable quantitative à représenter sous forme de courbe de densité.

shade : un booléen indiquant si la région sous la courbe de densité doit être remplie d'une couleur.

vertical : un booléen indiquant si la courbe de densité doit être tracée verticalement ou horizontalement.

kernel : le noyau à utiliser pour estimer la fonction de densité. Les options courantes sont "gau", "cos", "biw", "epa", "tri", "triw".

bw : la largeur de la fenêtre de noyau utilisée pour estimer la fonction de densité.

cut : la quantité de découpe des extrémités de la courbe de densité.

clip : les limites de l'axe pour lesquelles la courbe de densité doit être tracée.

legend : un booléen indiquant si une légende doit être ajoutée au graphique.

cumulative : un booléen indiquant si la courbe de densité cumulative doit être tracée.

color : la couleur de la courbe de densité.

Il est important de noter que kdeplot offre de nombreuses autres options pour personnaliser la représentation de la courbe de densité. Vous pouvez trouver la documentation complète pour la fonction kdeplot sur le site web de Seaborn.

```python
# Don't worry about understanding this code!
# It's just for the diagram below
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

#Create dataset
dataset = np.random.randn(25)

# Create another rugplot
sns.rugplot(dataset);

# Set up the x-axis for the plot
x_min = dataset.min() - 2
x_max = dataset.max() + 2

# 100 equally spaced points from x_min to x_max
x_axis = np.linspace(x_min,x_max,100)

# Set up the bandwidth, for info on this:
url = 'http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth'

bandwidth = ((4*dataset.std()**5)/(3*len(dataset)))**.2


# Create an empty kernel list
kernel_list = []

# Plot each basis function
for data_point in dataset:
    
    # Create a kernel for each point and append to list
    kernel = stats.norm(data_point,bandwidth).pdf(x_axis)
    kernel_list.append(kernel)
    
    #Scale for plotting
    kernel = kernel / kernel.max()
    kernel = kernel * .4
    plt.plot(x_axis,kernel,color = 'grey',alpha=0.5)

plt.ylim(0,1)
```

```python
# To get the kde plot we can sum these basis functions.

# Plot the sum of the basis function
sum_of_kde = np.sum(kernel_list,axis=0)

# Plot figure
fig = plt.plot(x_axis,sum_of_kde,color='indianred')

# Add the initial rugplot
sns.rugplot(dataset,c = 'indianred')

# Get rid of y-tick marks
plt.yticks([])

# Set title
plt.suptitle("Sum of the Basis Functions")
```
