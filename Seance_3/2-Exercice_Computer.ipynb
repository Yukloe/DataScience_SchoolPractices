Q4 Représenter sur price l'impact de ces couples de variables
            - speed et hd
            - hd et screen
            - speed et premium
            - hd et premium

1) Speed et HD

```python
sns.scatterplot(x= 'speed', y='hd', hue='price', data= df)
```

Plus on monte, plus nos points deviennent clairs, et donc augmentent de prix. De même pour la variable speed. Plus la vitesse augmente (vers la droite), plus notre prix augmente.

```python
sns.scatterplot(x= 'screen', y='hd', hue='price', data= df)
```

Le résultat est le même que précedemment.

```python
sns.scatterplot(x= 'premium', y='speed', hue='price', data= df)
```

La colonne yes est davantage clair que la colonne non. On voit donc que premium a également un effet.

```python
sns.scatterplot(x= 'premium', y='hd', hue='price', data= df)
```

Ici, c'est moins évident. Le prix augmente, mais surtout en montant. Donc en améliorant la qualité HD. Il est probable que HD ait davantage d'effet sur la variable prix.

```python
sns.relplot(x='ram', y='price', hue='premium', size='screen', style='cd', data=df)
```

On observe que premium=yes n'est pas tellement plus haut que non. Au contraire, beaucoup de PC premium sont même moins chère.
On observe de manière évidente qu'une RAM élevée augmente le prix. C'est donc un facteur à garder.
En moyenne, les points semblent plus gros au fur et à mesure que l'on monte sur l'échelle des prix. La taille de l'écran est donc un facteur à prendre en compte.
Les CD semblent plutôt dispersés. Son effet a donc moins d'impact sur le prix.

En conclusion, les deux facteurs impactant le plus le prix d'un ordinateur est : la RAM et la taille de l'écran.

On ne peut pas non plus observer de réel différence. On ne va pas pouvoir en tirer de conclusion.
